## Advertising problem with sklearn

```python
 #Dataset
 import pandas as pnd
 import matplotlib.pyplot as plt
 from sklearn.model_selection import train_test_split
 from sklearn.linear_model import LinearRegression
 from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score 

 # Load the dataset
 dataset=pnd.read_csv("Advertising.csv",index_col=0)
 # Selection of the first three columns of the dataset: these are the input variables (features)
 X=dataset.iloc[:,:3]
 # Selection of the fourth column: this is the target variable to predict (here: Sales = Data)
 y=dataset.iloc[:,3]
 # Separation of the dataset into 2 parts (train and test)
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)

 # Model creation and training
 model=LinearRegression() 
 model.fit(X_train,y_train) 
 # Prediction on test data
 yhat=model.predict(X_test)


# Model parameters
bias= model.intercept_
w = model.coef_

print("Biais (intercept) :", bias)
print("Coefficients (w) :", w)

mae = mean_absolute_error(y_test, yhat)
mse = mean_squared_error(y_test, yhat)
r2 = r2_score(y_test, yhat)

print("Mean Absolute Error :", mae)
# Erreur Quadratique Moyenne
print("Root Mean Square Error :", mse)
print("R2 Score :", r2)

# Visualizing the results
plt.figure(figsize=(12, 6))
# Comparison of actual and predicted values
plt.subplot(1, 2, 1)
plt.scatter(y_test, yhat, label='Predictions',color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2, label='Reference Line')
plt.xlabel('Real Data')
plt.ylabel('Predicted Data')
plt.title('Actual vs. Predicted Data')
plt.legend()
```
